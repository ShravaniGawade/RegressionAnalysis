```python
for column in numeric_columns:
    median_value = train_data[column].median()
    train_data[column].fillna(median_value, inplace = True)
```

```python
for column in non_numeric_columns:
    mode_value = train_data[column].mode()[0]
    train_data[column].fillna(mode_value, inplace = True)
```

The mode() function in Pandas returns a Series containing the most frequent value(s) in the column.
mode()[0] extracts the first element of this Series, which is the most frequent value itself (the mode).

```python
numeric_columns = test_data.select_dtypes(include = [int, float]).columns
non_numeric_columns = test_data.select_dtypes(exclude = [int, float]).columns

for column in numeric_columns:
    median_value = test_data[column].median()
    test_data[column].fillna(median_value, inplace = True)
    
for column in non_numeric_columns:
    mode_value = test_data[column].mode()[0]
    test_data[column].fillna(mode_value, inplace = True)
```

Linear Regression: A basic model that assumes a linear relationship between features and the target.
Random Forest: A robust ensemble method that handles non-linearity and interactions between features well.
Lasso Regression: Adds regularization to Linear Regression, penalizing large coefficients to prevent overfitting and perform feature selection.
Ridge Regression: Similar to Lasso, but instead of shrinking coefficients to zero, it reduces their size, making the model less sensitive to multicollinearity.

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

# Prepare the data
X = train_data.drop('target', axis=1)
Y = train_data['target']

# Split the data into training and validation sets
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.2, random_state=42)

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)

# Initialize and train the models
linear_model = LinearRegression()
linear_model.fit(X_train_scaled, Y_train)
y_pred_val_linear = linear_model.predict(X_val_scaled)

lasso_model = Lasso(alpha=0.1)
lasso_model.fit(X_train_scaled, Y_train)
y_pred_val_lasso = lasso_model.predict(X_val_scaled)

ridge_model = Ridge(alpha=0.1)
ridge_model.fit(X_train_scaled, Y_train)
y_pred_val_ridge = ridge_model.predict(X_val_scaled)

random_forest_model = RandomForestRegressor(n_estimators=100, random_state=42)
random_forest_model.fit(X_train, Y_train)
y_pred_val_random_forest = random_forest_model.predict(X_val)

# Evaluate the models on the validation set
print("Linear Regression Validation Set:")
print("Validation MSE:", mean_squared_error(Y_val, y_pred_val_linear))
print("Validation R2 Score:", r2_score(Y_val, y_pred_val_linear))

print("\nLasso Regression Validation Set:")
print("Validation MSE:", mean_squared_error(Y_val, y_pred_val_lasso))
print("Validation R2 Score:", r2_score(Y_val, y_pred_val_lasso))

print("\nRidge Regression Validation Set:")
print("Validation MSE:", mean_squared_error(Y_val, y_pred_val_ridge))
print("Validation R2 Score:", r2_score(Y_val, y_pred_val_ridge))

print("\nRandom Forest Validation Set:")
print("Validation MSE:", mean_squared_error(Y_val, y_pred_val_random_forest))
print("Validation R2 Score:", r2_score(Y_val, y_pred_val_random_forest))
```

Why Scaling is Not Required for Random Forest:-
Tree-Based Methods: Random Forest is an ensemble of decision trees, and tree-based algorithms do not rely on the distance between data points. Instead, they split the data based on feature values, which means the model is invariant to the scale of the features.

No Assumptions About Data:-
Unlike linear models (like Linear Regression, Lasso, and Ridge), Random Forest does not assume a particular distribution of the data, so features do not need to be normalized or standardized.

fit_transform: Used on the training data to both fit the scaler and transform the data based on the training statistics.
transform: Used on the test data to transform it using the same scaling parameters obtained from the training data.

Mean Squared Error (MSE)
Definition: MSE measures the average squared difference between the actual and predicted values. Lower MSE values indicate better model performance because the predictions are closer to the actual values.

Linear Regression: MSE = 12.68
Lasso Regression: MSE = 12.23
Ridge Regression: MSE = 12.68
Random Forest: MSE = 11.80

The Random Forest model has the lowest MSE, suggesting it has the best predictive accuracy among the models. The Lasso Regression model also performs better than Linear and Ridge Regression in terms of MSE.

R-squared (R²) Score
Definition: R² measures the proportion of the variance in the dependent variable that is predictable from the independent variables. R² ranges from 0 to 1, with higher values indicating better model performance. An R² of 0 means the model does not explain any variability, while an R² of 1 means the model explains all the variability.

Linear Regression: R² = 0.55
Lasso Regression: R² = 0.57
Ridge Regression: R² = 0.55
Random Forest: R² = 0.58

The Random Forest model has the highest R² score, meaning it explains the largest proportion of variance in the target variable. Lasso Regression also performs better than Linear and Ridge Regression in terms of R².
